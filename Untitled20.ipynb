```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import json
from ase import Atoms
from ase.build import fcc111, add_adsorbate
from ase.optimize import FIRE
from ase.constraints import FixAtoms
from ase.io import write
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from datetime import datetime
import sys
from chgnet.model.model import CHGNet
from chgnet.model.dynamics import CHGNetCalculator

print(f"╔{'═'*70}╗")
print(f"║{'HgO/Au(111) Adsorption Study - CHGNet Potential':^70}║")
print(f"╠{'═'*70}╣")
print(f"║ Python: {sys.version:59} ║")
print(f"║ Time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S'):^70} ║")
print(f"╚{'═'*70}╝")

print("\n[1] Loading CHGNet Model...")
model = CHGNet.load()
calc = CHGNetCalculator(model=model)
print("✓ Model initialized")
```

```python
def relax_structure(atoms, label, fix_bottom=True, fmax=0.02, max_steps=500):
    atoms.calc = calc
    if fix_bottom and len(atoms) > 10:
        z_coords = atoms.positions[:, 2]
        mask = z_coords < np.percentile(z_coords, 33)
        atoms.set_constraint(FixAtoms(mask=mask))

    dyn = FIRE(atoms, logfile=None)
    e_init = atoms.get_potential_energy()
    dyn.run(fmax=fmax, steps=max_steps)
    e_final = atoms.get_potential_energy()

    print(f"  {label}: {e_init:.4f} → {e_final:.4f} eV ({dyn.get_number_of_steps()} steps)")
    write(f"{label}.xyz", atoms)

    return {'energy': e_final, 'atoms': atoms.copy(), 'label': label}

def calc_e_ads(e_total, e_slab, e_mol):
    return e_total - (e_slab + e_mol)

def analyze_structure_detailed(atoms, label="Structure"):
    au_indices = [i for i, a in enumerate(atoms) if a.symbol == 'Au']
    o_indices = [i for i, a in enumerate(atoms) if a.symbol == 'O']
    hg_indices = [i for i, a in enumerate(atoms) if a.symbol == 'Hg']

    au_positions = atoms.positions[au_indices]
    o_position = atoms.positions[o_indices[0]] if o_indices else None
    hg_position = atoms.positions[hg_indices[0]] if hg_indices else None

    z_surface = np.max(au_positions[:, 2])
    o_height = o_position[2] - z_surface if o_position is not None else None
    hg_height = hg_position[2] - z_surface if hg_position is not None else None

    results = {}
    if o_position is not None and hg_position is not None:
        bond_vector = hg_position - o_position
        bond_length = np.linalg.norm(bond_vector)
        bond_horizontal = np.linalg.norm(bond_vector[:2])
        bond_vertical = abs(bond_vector[2])
        tilt_angle = np.arctan2(bond_horizontal, bond_vertical) * 180 / np.pi

        distances_to_o = [np.linalg.norm(pos - o_position) for pos in au_positions]
        distances_to_hg = [np.linalg.norm(pos - hg_position) for pos in au_positions]

        d_au_o = np.min(distances_to_o)
        d_au_hg = np.min(distances_to_hg)

        results = {
            'bond_length': bond_length,
            'bond_change': bond_length - 2.01,
            'o_height': o_height,
            'hg_height': hg_height,
            'tilt_angle': tilt_angle,
            'd_au_o': d_au_o,
            'd_au_hg': d_au_hg,
            'vertical': bond_horizontal < 0.1
        }

        print(f"\n{'='*60}")
        print(f"STRUCTURAL ANALYSIS: {label}")
        print(f"{'='*60}")
        print(f"Hg-O bond length:     {bond_length:.3f} Å  (gas: 2.010 Å)")
        print(f"  Change from gas:    {bond_length - 2.010:+.3f} Å")
        print(f"O height:             {o_height:.3f} Å")
        print(f"Hg height:            {hg_height:.3f} Å")
        print(f"Bond tilt angle:      {tilt_angle:.1f}°")
        print(f"Nearest Au-O:         {d_au_o:.3f} Å")
        print(f"Nearest Au-Hg:        {d_au_hg:.3f} Å")
        print(f"{'='*60}")

    return results

def create_comprehensive_dashboard(results_data, best_atoms=None):
    fig = make_subplots(
        rows=3, cols=2,
        subplot_titles=(
            'Adsorption Energies by Site',
            'Orientation Comparison',
            'Height Dependence',
            'Top 10 Configurations',
            'Structural Parameters (Best)',
            'Energy Distribution'
        ),
        specs=[
            [{"type": "bar"}, {"type": "bar"}],
            [{"type": "scatter"}, {"type": "bar"}],
            [{"type": "bar"}, {"type": "histogram"}]
        ],
        vertical_spacing=0.1,
        horizontal_spacing=0.1
    )

    sites = [r['site'] for r in results_data]
    energies = [r['energy_ads'] for r in results_data]
    orientations = [r['orientation'] for r in results_data]
    heights = [r.get('height', 2.5) for r in results_data]

    site_avg = {s: np.mean([e for e, si in zip(energies, sites) if si == s]) for s in set(sites)}
    fig.add_trace(
        go.Bar(x=list(site_avg.keys()), y=list(site_avg.values()),
               marker_color=['#3498db', '#2ecc71', '#e74c3c', '#f39c12']),
        row=1, col=1
    )

    orient_avg = {o: np.mean([e for e, orien in zip(energies, orientations) if orien == o]) for o in set(orientations)}
    fig.add_trace(
        go.Bar(x=list(orient_avg.keys()), y=list(orient_avg.values()),
               marker_color=['#e67e22', '#9b59b6']),
        row=1, col=2
    )

    fig.add_trace(
        go.Scatter(x=heights, y=energies, mode='markers',
                  marker=dict(size=12, color=energies, colorscale='RdYlGn_r',
                             showscale=False, line=dict(width=1, color='black'))),
        row=2, col=1
    )

    sorted_results = sorted(results_data, key=lambda x: x['energy_ads'])[:10]
    top_labels = [f"{r['site'][:3]}_{r['orientation'][:3]}" for r in sorted_results]
    top_energies = [r['energy_ads'] for r in sorted_results]
    colors_top = ['#2ecc71' if e < -1.8 else '#f39c12' if e < -1.5 else '#e74c3c' for e in top_energies]

    fig.add_trace(
        go.Bar(x=top_labels, y=top_energies, marker_color=colors_top,
               text=[f"{e:.2f}" for e in top_energies], textposition='auto'),
        row=2, col=2
    )

    if best_atoms is not None:
        analysis = analyze_structure_detailed(best_atoms, "Best")
        if analysis:
            params = ['Bond (Å)', 'O Height (Å)', 'Hg Height (Å)', 'Tilt (°)', 'Au-O (Å)']
            values = [analysis['bond_length'], analysis['o_height'],
                     analysis['hg_height'], analysis['tilt_angle'], analysis['d_au_o']]
            fig.add_trace(
                go.Bar(x=params, y=values, marker_color='#1abc9c',
                       text=[f"{v:.2f}" for v in values], textposition='auto'),
                row=3, col=1
            )

    fig.add_trace(
        go.Histogram(x=energies, nbinsx=12, marker_color='#34495e'),
        row=3, col=2
    )

    fig.update_layout(
        height=1200, width=1400,
        title_text="<b>Comprehensive HgO/Au(111) Analysis Dashboard</b>",
        showlegend=False,
        paper_bgcolor='white',
        plot_bgcolor='white'
    )

    return fig
```

```python
print("\n[2] Reference HgO Molecule")
print("=" * 70)

hgo = Atoms('HgO', positions=[[0, 0, 0], [0, 0, 2.05]])
hgo.center(vacuum=10.0)
hgo.calc = calc

res_hgo = relax_structure(hgo, "HgO_gas", fix_bottom=False, fmax=0.01)
e_hgo = res_hgo['energy']
print(f"  Reference energy: {e_hgo:.4f} eV")
```

```python
print("\n[3] Building Au(111) Surface")
print("=" * 70)

slab = fcc111('Au', size=(3, 3, 4), vacuum=15.0, periodic=True)
print(f"  Atoms: {len(slab)} | Cell: {slab.get_cell()[0,0]:.2f} × {slab.get_cell()[1,1]:.2f} Å")

print("\n[4] Relaxing Surface")
print("=" * 70)

res_slab = relax_structure(slab.copy(), "Au_slab", fix_bottom=True, fmax=0.03)
e_slab = res_slab['energy']
slab_relaxed = res_slab['atoms']
print(f"  Surface energy: {e_slab:.4f} eV")
```

```python
print("\n[5] Adsorption Study")
print("=" * 70)

def add_hgo(slab, site, orient, h_hg=2.5, h_o=4.5):
    s = slab.copy()
    cell = s.get_cell()
    x_c, y_c = cell[0,0]/2, cell[1,1]/2
    d = 1.4

    sites = {
        'ontop': (x_c, y_c),
        'bridge': (x_c + d, y_c),
        'fcc': (x_c + d/2, y_c + d*0.866),
        'hcp': (x_c - d/2, y_c + d*0.866)
    }
    pos = sites.get(site, (x_c, y_c))

    if orient == 'Hg_down':
        add_adsorbate(s, 'Hg', h_hg, pos)
        add_adsorbate(s, 'O', h_o, pos)
    else:
        add_adsorbate(s, 'O', h_hg, pos)
        add_adsorbate(s, 'Hg', h_o, pos)

    return s

sites = ['ontop', 'bridge', 'fcc', 'hcp']
orients = ['Hg_down', 'O_down']
heights = [2.0, 2.5, 3.0]

results_matrix = []

for site in sites:
    for ori in orients:
        for h in heights:
            label = f"{site}_{ori}_h{h}"
            try:
                sys = add_hgo(slab_relaxed, site, ori, h, h+2.0)
                res = relax_structure(sys, label, fix_bottom=True, fmax=0.05, max_steps=300)
                e_ads = calc_e_ads(res['energy'], e_slab, e_hgo)

                results_matrix.append({
                    'site': site,
                    'orientation': ori,
                    'height': h,
                    'energy_total': res['energy'],
                    'energy_ads': e_ads,
                    'atoms': res['atoms'],
                    'label': label
                })

                status = "STABLE" if e_ads < 0 else "unstable"
                print(f"    {status}: E_ads = {e_ads:.3f} eV")

            except Exception as e:
                print(f"    ✗ {label}: {str(e)[:40]}")

print(f"\n  Completed: {len(results_matrix)} configurations")
```

```python
print("\n[6] Results Analysis")
print("=" * 70)

results_matrix.sort(key=lambda x: x['energy_ads'])

print(f"\n{'Rank':<5} {'Configuration':<25} {'E_ads (eV)':<12}")
print("-" * 45)
for i, r in enumerate(results_matrix[:10]):
    print(f"{i+1:<5} {r['label']:<25} {r['energy_ads']:<12.3f}")

best_result = results_matrix[0]
best_atoms = best_result['atoms']

print(f"\n  >>> GLOBAL MINIMUM: {best_result['label']}")
print(f"      E_ads = {best_result['energy_ads']:.4f} eV")

detailed_analysis = analyze_structure_detailed(best_atoms, f"Best: {best_result['site']}_{best_result['orientation']}")
```

```python
print("\n[7] Generating Dashboard")
print("=" * 70)

dash_fig = create_comprehensive_dashboard(results_matrix, best_atoms)
dash_fig.show()
```

```python
print("\n[8] Interactive 3D Visualization")
print("=" * 70)

pos = best_atoms.positions
sym = best_atoms.get_chemical_symbols()

colors = {'Au': '#FFD700', 'Hg': '#808080', 'O': '#FF4500'}
sizes = {'Au': 7, 'Hg': 8, 'O': 6}

fig_3d = go.Figure()

for elem in ['Au', 'Hg', 'O']:
    indices = [i for i, s in enumerate(sym) if s == elem]
    if indices:
        fig_3d.add_trace(go.Scatter3d(
            x=pos[indices, 0], y=pos[indices, 1], z=pos[indices, 2],
            mode='markers',
            name=elem,
            marker=dict(size=sizes[elem], color=colors[elem], opacity=0.9,
                       line=dict(color='black', width=1))
        ))

for i in range(len(best_atoms)):
    for j in range(i+1, len(best_atoms)):
        d = np.linalg.norm(pos[i] - pos[j])
        pair = tuple(sorted([sym[i], sym[j]]))
        cutoff = {('Au', 'Au'): 3.0, ('Hg', 'O'): 2.5, ('Au', 'Hg'): 3.2, ('Au', 'O'): 2.8}.get(pair, 0)
        if cutoff > 0 and d < cutoff:
            fig_3d.add_trace(go.Scatter3d(
                x=[pos[i,0], pos[j,0]], y=[pos[i,1], pos[j,1]], z=[pos[i,2], pos[j,2]],
                mode='lines',
                line=dict(color='gray', width=3),
                showlegend=False
            ))

fig_3d.update_layout(
    title=f"Optimal: {best_result['label']} | E_ads = {best_result['energy_ads']:.3f} eV",
    scene=dict(xaxis=dict(visible=False), yaxis=dict(visible=False), zaxis=dict(visible=False)),
    width=900, height=700
)
fig_3d.show()
```

```python
print("\n[9] Saving Results")
print("=" * 70)

df = pd.DataFrame([{
    'configuration': r['label'],
    'site': r['site'],
    'orientation': r['orientation'],
    'height_A': r['height'],
    'E_ads_eV': r['energy_ads'],
    'E_total_eV': r['energy_total']
} for r in results_matrix])

df.to_csv('adsorption_results.csv', index=False)
write('best_structure.xyz', best_atoms)

with open("final_results.json", "w") as f:
    json.dump({
        'best_config': {
            'site': best_result['site'],
            'orientation': best_result['orientation'],
            'height': best_result['height'],
            'energy_ads': float(best_result['energy_ads']),
            'structural_params': {k: float(v) for k, v in detailed_analysis.items()
                                if isinstance(v, (int, float))}
        },
        'all_results': [{'site': r['site'], 'orientation': r['orientation'],
                        'height': r['height'], 'energy_ads': float(r['energy_ads'])}
                       for r in results_matrix]
    }, f, indent=2)

print("✓ Saved: adsorption_results.csv")
print("✓ Saved: best_structure.xyz")
print("✓ Saved: final_results.json")

print("\n" + "="*70)
print("STUDY COMPLETE")
print("="*70)
print(f"Configurations: {len(results_matrix)}")
print(f"Stable (E<0): {len([r for r in results_matrix if r['energy_ads'] < 0])}")
print(f"Global Minimum: {best_result['label']}")
print(f"E_ads = {best_result['energy_ads']:.4f} eV")
if detailed_analysis:
    print(f"Hg-O Bond = {detailed_analysis['bond_length']:.3f} Å")
    print(f"Tilt Angle = {detailed_analysis['tilt_angle']:.1f}°")
print("="*70)
```
