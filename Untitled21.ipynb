```python
#
# Simple program on adsorption
#
# From ASE test suite
#
from math import cos, pi, sin

from ase import Atoms                    # لإنشاء الذرات والجزيئات
from ase.build import add_adsorbate, fcc111  # لبناء السطح وإضافة المادة الممتزة
from ase.calculators.emt import EMT      # حاسبة الطاقة (Effective Medium Theory)
from ase.constraints import FixLinearTriatomic  # لتثبيت شكل جزيء CO2
from ase.optimize import BFGS            # خوارزمية التحسين الهندسي
from ase.visualize import view           # للعرض المرئي

# ============================================
# الخطوة 1: بناء جزيء CO2 (ثاني أكسيد الكربون)
# ============================================

# حساب المواقع الهندسية لذرات CO2
# CO2 جزيء خطي: O=C=O مع زاوية 180 درجة
# لكن هنا بيستخدم زاوية 134.3° (شكل معين)
zpos = cos(134.3 / 2.0 * pi / 180.0) * 1.197
xpos = sin(134.3 / 2.0 * pi / 180.0) * 1.19

print('xpos=', xpos, ' zpos=', zpos)

# إنشاء جزيء CO2
# الذرة الأولى: كربون (C)
# الذرتين 2 و 3: أكسجين (O, O)
co2 = Atoms('COO', positions=[
    (-xpos + 1.2, 0, -zpos),      # الكربون (في المنتصف)
    (-xpos + 1.2, -1.1, -zpos),   # أكسجين أول
    (-xpos + 1.2, +1.1, -zpos)    # أكسجين ثاني
])

# ============================================
# الخطوة 2: بناء سطح الذهب Au(111)
# ============================================

# fcc111 = سطح مكعب وجهي مركزي في الاتجاه [111]
# size=(6, 6, 4) = 6×6 ذرات في السطح، 4 طبقات
# vacuum=20 أنجستروم (فراغ فوق السطح)
slab = fcc111('Au', size=(6, 6, 4), vacuum=2 * 10, orthogonal=True)
slab.center()  # توسيط السطح

# ============================================
# الخطوة 3: إضافة CO2 على السطح (الامتزاز)
# ============================================

# pos_shift = موقع الإضافة على السطح
pos_shift = 5.0
add_adsorbate(slab, co2, height=1.5, position=(pos_shift, pos_shift))
# height=1.5 أنجستروم فوق السطح

# عرض النظام (يفتح نافذة رسومية)
view(slab)

# ============================================
# الخطوة 4: إعدادات الدورية والحاسبة
# ============================================

# Periodic Boundary Conditions: دورية في x,y فقط (لا في z)
slab.set_pbc((True, True, False))

# قياس المسافات الأصلية في CO2 (قبل التحسين)
d0 = co2.get_distance(-3, -2)  # C-O1
d1 = co2.get_distance(-3, -1)  # C-O2
d2 = co2.get_distance(-2, -1)  # O1-O2
print("CO2 interatomic distances:", d0, d1, d2)

# اختيار الحاسبة: EMT (نظرية الوسط الفعّال)
# EMT حاسبة تقريبية سريعة للمعادن
calc = EMT()
slab.calc = calc

# ضبط المواقع في الخلية الدورية
slab.set_scaled_positions(slab.get_scaled_positions() % 1.0)

# ============================================
# الخطوة 5: القيود والتحسين
# ============================================

# تثبيت شكل CO2 الخطي أثناء التحسين
# (-2, -3, -1) = (O1, C, O2) - الفهارس السلبية تعني من النهاية
constraint = FixLinearTriatomic(triples=[(-2, -3, -1)])
slab.set_constraint(constraint)

print("Number of degrees of freedom =", slab.get_number_of_degrees_of_freedom())

# ============================================
# الخطوة 6: التحسين الهندسي (Geometry Optimization)
# ============================================

print("running CO2@Au(111) geometry optimization:")

# BFGS = خوارزمية تحسين شائعة
# trajectory='relaxation.traj' = حفظ النتائج في ملف
with BFGS(slab, trajectory='relaxation.traj') as dyn:
    dyn.run(fmax=0.05)  # fmax=قوة القصوى المسموحة (eV/Å)

# ============================================
# التحقق: هل CO2 حافظ على شكله؟
# ============================================

assert abs(slab.get_distance(-3, -2, mic=1) - d0) < 1e-9
assert abs(slab.get_distance(-3, -1, mic=1) - d1) < 1e-9
assert abs(slab.get_distance(-2, -1, mic=1) - d2) < 1e-9

print("assertion on CO2 interatomic distances:",
      abs(slab.get_distance(-3, -2, mic=1) - d0),
      abs(slab.get_distance(-3, -1, mic=1) - d1),
      abs(slab.get_distance(-2, -1, mic=1) - d2))
```

```python
#
# CO2 Adsorption on Metals - Enhanced Version
#
from math import cos, pi, sin
from ase import Atoms
from ase.build import add_adsorbate, fcc111
from ase.calculators.emt import EMT
from ase.constraints import FixLinearTriatomic
from ase.optimize import BFGS
import matplotlib.pyplot as plt
from ase.io import read

def run_adsorption(metal, height=1.5, size=(4, 4, 3)):
    """تشغيل امتزاز CO2 على معدن معين"""

    # بناء CO2
    zpos = cos(134.3 / 2.0 * pi / 180.0) * 1.197
    xpos = sin(134.3 / 2.0 * pi / 180.0) * 1.19
    co2 = Atoms('COO', positions=[
        (-xpos + 1.2, 0, -zpos),
        (-xpos + 1.2, -1.1, -zpos),
        (-xpos + 1.2, +1.1, -zpos)
    ])

    # طاقة CO2 المنفصل
    co2.calc = EMT()
    E_co2 = co2.get_potential_energy()

    # بناء السطح النظيف
    slab_clean = fcc111(metal, size=size, vacuum=15, orthogonal=True)
    slab_clean.calc = EMT()
    E_slab = slab_clean.get_potential_energy()

    # بناء نظام الامتزاز
    slab = fcc111(metal, size=size, vacuum=15, orthogonal=True)
    add_adsorbate(slab, co2, height, 'fcc')
    slab.calc = EMT()

    # قيود
    constraint = FixLinearTriatomic(triples=[(-2, -3, -1)])
    slab.set_constraint(constraint)

    # تحسين
    with BFGS(slab, trajectory=f'{metal}_co2.traj') as dyn:
        dyn.run(fmax=0.05)

    E_final = slab.get_potential_energy()
    E_ads = E_final - E_slab - E_co2

    return {
        'metal': metal,
        'E_adsorption': E_ads,
        'E_final': E_final,
        'E_slab': E_slab,
        'E_co2': E_co2
    }

# تشغيل على معادن مختلفة
metals = ['Au', 'Pt', 'Cu', 'Ag']
results = []

for metal in metals:
    print(f"\n{'='*40}")
    print(f"Running {metal}...")
    result = run_adsorption(metal)
    results.append(result)
    print(f"Adsorption Energy: {result['E_adsorption']:.3f} eV")

# جدول مقارنة
print(f"\n{'='*50}")
print("SUMMARY - Adsorption Energies")
print(f"{'Metal':<10} {'E_ads (eV)':<15} {'Strength'}")
print("-" * 50)
for r in results:
    strength = "Strong" if r['E_adsorption'] < -0.5 else "Weak" if r['E_adsorption'] < 0 else "Repulsive"
    print(f"{r['metal']:<10} {r['E_adsorption']:<15.3f} {strength}")

# رسم بياني
metals_list = [r['metal'] for r in results]
energies_list = [r['E_adsorption'] for r in results]

plt.figure(figsize=(8, 5))
plt.bar(metals_list, energies_list, color=['gold', 'gray', 'orange', 'silver'])
plt.axhline(y=0, color='r', linestyle='--', label='No interaction')
plt.ylabel('Adsorption Energy (eV)')
plt.title('CO2 Adsorption on Different Metals')
plt.legend()
plt.savefig('adsorption_comparison.png', dpi=150)
plt.show()
```
